```python
import pandas as pd

# 1. 샘플 데이터 준비
# 데이터를 포함한 CSV 파일을 불러온다고 가정
# 데이터 예시: 'date', 'company', 'market_cap', 'price' 열 포함
data = pd.read_csv('C:/Users/DaBin/Desktop/대학원/2024-2/통계계산특론/PROJECT/data/최종데이터/top10_market_cap.csv')
firm = pd.read_csv('C:/Users/DaBin/Desktop/대학원/2024-2/통계계산특론/PROJECT/data/firm_new.csv')

# 투자 결과를 저장할 데이터프레임 초기화
results = []

# 2. 매월 Top 10 회사 필터링 및 투자 수행
start_year = 2016
start_month = 12

while start_year <= 2022 :
    
    
        
    top10 = data.loc[(data['month'] == start_month) & (data['year'] == start_year),'Ticker']
    filtered_firm_data = firm[firm['ticker'].isin(top10)]

    firm_month = start_month + 1
    
    if firm_month >= 13 :
        firm_month = start_month -12
        firm_year = start_year + 1
        
    firm10 = filtered_firm_data.loc[(filtered_firm_data['month'] == firm_month ) & (filtered_firm_data['year'] == firm_year),['ticker','ret']]
   
   
    # 투자 비중 계산 (균등 투자와 시가총액 비례 투자)
    total_market_cap = top10['Market_Cap'].sum()
    top10['equal_weight'] = 1 / len(top10)
    top10['cap_weight'] = top10['Market_Cap'] / total_market_cap

    # 투자 금액 설정
    investment = 10000  
    # 각 투자 방법에 따른 종목당 금액 계산
    top10['equal_investment'] = top10['equal_weight'] * investment
    top10['cap_investment'] = top10['cap_weight'] * investment

    # 수익 계산
    
     
    top10['equal_return'] = top10['equal_investment'] * (1+firm10['ret'])
    top10['cap_return'] = top10['cap_investment'] * (1+firm10['ret'])
    
    
      
    # 결과 저장
    results.append({
        'year' : firm_year,
        'month' : firm_month,
        'equal_return': top10['equal_return'].sum(),
        'cap_return': top10['cap_return'].sum(),
        'equal_return_rate' : 
    
    })

    # 다음 달로 이동
    current_date += pd.DateOffset(months=1)

# 3. 결과 데이터프레임 생성 및 출력
results_df = pd.DataFrame(results)

print("투자 결과")
print(results_df)

# 4. 결과 시각화 (수익률 비교)
import matplotlib.pyplot as plt

results_df['cumulative_equal_return'] = (1 + results_df['equal_return']).cumprod()
results_df['cumulative_cap_return'] = (1 + results_df['cap_return']).cumprod()

plt.figure(figsize=(12, 6))
plt.plot(results_df['date'], results_df['cumulative_equal_return'], label='Equal Weight')
plt.plot(results_df['date'], results_df['cumulative_cap_return'], label='Cap Weight')
plt.title('Cumulative Return Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.show()
```
