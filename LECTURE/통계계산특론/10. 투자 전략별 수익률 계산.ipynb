### 균등 투자

```python
import pandas as pd
import numpy as np

# 초기 투자 금액
initial_investment = 10000

# 투자 시뮬레이션 데이터프레임 생성
investment_results = []

# 2017년 1월부터 2022년 3월까지 월별로 루프
start_year = 2017
start_month = 1
end_year = 2022
end_month = 3

# 투자 금액 초기화
current_investment = initial_investment

# 월별 반복 루프
for year in range(start_year, end_year + 1):
    for month in range(1, 13):
        # 종료 조건
        if year == end_year and month > end_month:
            break
        
        # data에서 해당 연도와 월의 상위 티커 필터링
        filtered_data = data[(data['year'] == year) & (data['month'] == month)]

        # 해당 월에 투자할 티커 목록 가져오기
        tickers = filtered_data['Ticker'].tolist()

        # filtered_firm_data에서 해당 연도와 월에 대한 수익률 데이터 필터링
        returns_data = filtered_firm_data[
            (filtered_firm_data['year'] == year) &
            (filtered_firm_data['month'] == month) &
            (filtered_firm_data['ticker'].isin(tickers))
        ]

        # 만약 해당 월에 투자할 티커 데이터가 없으면 넘어감
        if returns_data.empty:
            investment_results.append({
                'year': year,
                'month': month,
                'investment': current_investment,
                'return_rate': None,
                'updated_investment': current_investment,
                'num_companies': 0  # 투자한 회사 수 추가
            })
            continue

        # 투자 금액을 티커 개수만큼 균등 분배
        num_tickers = len(returns_data['ticker'].unique())  # 투자한 회사 수
        investment_per_ticker = current_investment / num_tickers

        # 각 티커의 수익률 적용 후 총 투자 금액 계산
        returns_data['investment_value'] = investment_per_ticker * (1 + returns_data['ret'])
        updated_investment = returns_data['investment_value'].sum()

        # 수익률 계산
        return_rate = (updated_investment - current_investment) / current_investment * 100

        # 결과 저장
        investment_results.append({
            'year': year,
            'month': month,
            'investment': current_investment,
            'return_rate': return_rate,
            'updated_investment': updated_investment,
            'num_companies': num_tickers  # 투자한 회사 수 추가
        })

        # 현재 투자 금액 업데이트
        current_investment = updated_investment

# 결과를 데이터프레임으로 변환
sp500_investment_results_df = pd.DataFrame(investment_results)

# 결과 출력
sp500_investment_results_df
```

### 시총 비례 투자

```python
import pandas as pd
import numpy as np

# 초기 투자 금액
initial_investment = 10000

# 투자 시뮬레이션 데이터프레임 생성
investment_results = []

# 2017년 1월부터 2022년 3월까지 월별로 루프
start_year = 2017
start_month = 1
end_year = 2022
end_month = 3

# 투자 금액 초기화
current_investment = initial_investment

# 월별 반복 루프
for year in range(start_year, end_year + 1):
    for month in range(1, 13):
        # 종료 조건
        if year == end_year and month > end_month:
            break
        
        # data에서 해당 연도와 월의 상위 티커 필터링
        filtered_data = data[(data['year'] == year) & (data['month'] == month)]

        # filtered_firm_data에서 해당 연도와 월에 대한 수익률 데이터 필터링
        returns_data = filtered_firm_data[
            (filtered_firm_data['year'] == year) &
            (filtered_firm_data['month'] == month)
        ]

        # 두 데이터의 티커 교집합 필터링
        common_tickers = set(filtered_data['Ticker']).intersection(set(returns_data['ticker']))
        filtered_data = filtered_data[filtered_data['Ticker'].isin(common_tickers)]
        returns_data = returns_data[returns_data['ticker'].isin(common_tickers)]

        # 만약 교집합된 티커가 없으면 넘어감
        if filtered_data.empty or returns_data.empty:
            investment_results.append({
                'year': year,
                'month': month,
                'investment': current_investment,
                'return_rate': None,
                'updated_investment': current_investment,
                'num_companies': 0  # 투자한 회사 수
            })
            continue

        # 시가총액 비율로 가중치 계산
        filtered_data['weight'] = filtered_data['Market_Cap'] / filtered_data['Market_Cap'].sum()

        # 각 티커의 투자 금액 계산
        returns_data = returns_data.merge(filtered_data[['Ticker', 'weight']], left_on='ticker', right_on='Ticker', how='inner')
        returns_data['investment_value'] = current_investment * returns_data['weight'] * (1 + returns_data['ret'])

        # 총 투자 금액 계산
        updated_investment = returns_data['investment_value'].sum()

        # 수익률 계산
        return_rate = (updated_investment - current_investment) / current_investment * 100

        # 결과 저장
        investment_results.append({
            'year': year,
            'month': month,
            'investment': current_investment,
            'return_rate': return_rate,
            'updated_investment': updated_investment,
            'num_companies': len(common_tickers)  # 투자한 회사 수
        })

        # 현재 투자 금액 업데이트
        current_investment = updated_investment

# 결과를 데이터프레임으로 변환
sp500_capinvestment_results_df = pd.DataFrame(investment_results)

sp500_capinvestment_results_df
```
