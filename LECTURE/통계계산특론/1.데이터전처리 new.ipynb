```python
# 상관계수가 0.5이상인 변수 출력

import pandas as pd
import numpy as np

# 상관계수 계산
correlation_matrix = macro.corr()

# 상관계수 0.5 이상인 변수 쌍 추출
threshold = 0.8  # 기준값
correlated_vars = []

# 상삼각행렬 부분만 탐색하여 중복 제거
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) >= threshold:  # 절댓값으로 확인
            var1 = correlation_matrix.columns[i]
            var2 = correlation_matrix.columns[j]
            corr_value = correlation_matrix.iloc[i, j]
            correlated_vars.append((var1, var2, corr_value))

correlated_vars.sort()
print("상관계수가 0.8 이상인 변수 쌍:")
for var1, var2, corr_value in correlated_vars:
    print(f"{var1} - {var2}: {corr_value:.3f}")

```

# 3. 모델링

## 1) XGBoost

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
import xgboost as xgb


# 결과 저장 리스트
adj_results = [] 
r2_results = []

y=firm[['ret','year']]
 # Validation 연도별 반복 (2012년부터 2016년까지)
for validation_year in range(2012, 2017):
    # 1. Train 데이터: 1997 ~ (validation_year - 1)
    train_X = data[(data['year'] >= 1997) & (data['year'] <= validation_year - 1)].drop(columns=['year', 'month'])
    train_y = y[(y['year'] >= 1997) & (y['year'] <= validation_year - 1)]['ret']

    # 2. Validation 데이터: 해당 validation_year
    validation_X = data[data['year'] == validation_year].drop(columns=['year', 'month'])
    validation_y = y[y['year'] == validation_year]['ret']

    # 3. XGBoost 모델 정의
    model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
    model.fit(train_X, train_y)  # 모델 학습

    # 4. Validation 예측
    y_pred = model.predict(validation_X)

    # 5. 성능 측정 (Adjusted R-squared)
    r_squared = r2_score(validation_y, y_pred)
    n = len(validation_y)  # Number of samples
    k = train_X.shape[1]  # Number of predictors
    adjusted_r2 = 1 - ((1 - r_squared) * (n - 1)) / (n - k - 1)

    # 결과 저장
    r2_results.append(r_squared)
    adj_results.append(adjusted_r2)

# 각 데이터셋의 평균 Adjusted R-squared 계산
avg_adjusted_r2 = np.mean(adj_results)


# 출력
print("Final Results:")
print(avg_adjusted_r2)
print("R2 : ", np.mean(r2_results))
```

## 2) LGBM

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
import lightgbm as lgb
import re


# 결과 저장 리스트
adj_results = [] 
r2_results = []

y=firm[['ret','year']]
 # Validation 연도별 반복 (2012년부터 2016년까지)
for validation_year in range(2012, 2017):
    # 1. Train 데이터: 1997 ~ (validation_year - 1)
    train_X = data[(data['year'] >= 1997) & (data['year'] <= validation_year - 1)].drop(columns=['year', 'month'])
    train_y = y[(y['year'] >= 1997) & (y['year'] <= validation_year - 1)]['ret']

    # 2. Validation 데이터: 해당 validation_year
    validation_X = data[data['year'] == validation_year].drop(columns=['year', 'month'])
    validation_y = y[y['year'] == validation_year]['ret']
    
    model = lgb.LGBMRegressor(random_state = 42)
    train_X.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))

    # 데이터프레임 컬럼 이름
    columns = train_X.columns

    # 허용되지 않는 특수 문자 패턴 정의
    invalid_pattern = re.compile(r'[{}[\]:,\\/"\'\s\n\t]')

    # 문제 있는 컬럼 이름 탐지
    invalid_columns = [col for col in columns if invalid_pattern.search(col)]
    train_X.columns = train_X.columns.str.replace(r'[{}[\]:,\\/"\'\s\n\t]', '', regex=True)
    validation_X.columns = validation_X.columns.str.replace(r'[{}[\]:,\\/"\'\s\n\t]', '', regex=True)

    model.fit(train_X, train_y)  # 모델 학습

    # 4. Validation 예측
    y_pred = model.predict(validation_X)

    # 5. 성능 측정 (Adjusted R-squared)
    r_squared = r2_score(validation_y, y_pred)
    n = len(validation_y)  # Number of samples
    k = train_X.shape[1]  # Number of predictors
    adjusted_r2 = 1 - ((1 - r_squared) * (n - 1)) / (n - k - 1)

    # 결과 저장
    r2_results.append(r_squared)
    adj_results.append(adjusted_r2)

# 각 데이터셋의 평균 Adjusted R-squared 계산
avg_adjusted_r2 = np.mean(adj_results)


# 출력
print("Final Results:")
print(avg_adjusted_r2)
print("R2 : ", np.mean(r2_results))
```
