```python
# Component 변동성 반영 비율
print(pca.explained_variance_ratio_)
```

```python
macro = pd.read_csv('C:/Users/DaBin/Desktop/대학원/2024-2/통계계산특론/PROJECT/data/macroeconomic_97_22.csv')
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow.keras.layers import Dense, Dropout, LSTM, Activation
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import MSE
```

```python
macro = macro.drop('sasdate',axis = 1)
macro = np.reshape(macro,(323,1,124)) 
 
# 모델 정의
model = Sequential([
    LSTM(units=64, activation='relu', input_shape=(1,124)),  # 시퀀스 요약
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='linear')  # 차원 축소: 최종 출력 차원 10
])

# 모델 컴파일
model.compile(optimizer='adam', loss='mse')

# 출력 확인
output = model.predict(macro)
print("Input shape:", macro.shape)  
print("Output shape:", output.shape)
```

```python
lstm_10 = pd.DataFrame(output, 
                      columns = [f"LSTM{num+1}" for num in range(10)])
```

```python
def analyze_feature_importance(model, X, num_samples=1000):

    baseline_pred = model.predict(X)

    importances = []

    for feature_idx in range(X.shape[2]):

        X_perturbed = X.copy()
        X_perturbed[:, :, feature_idx] = np.random.permutation(X_perturbed[:, :, feature_idx])


        perturbed_pred = model.predict(X_perturbed)


        importance = np.mean(np.abs(perturbed_pred - baseline_pred))
        importances.append(importance)

    return importances


import pandas as pd
import numpy as np
import tensorflow as tf

from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Dropout, LSTM, Activation
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import MSE


#sasdate column drop

macro = macro.drop('sasdate',axis = 1)


X_train =  macro.drop('S&P 500', axis = 1)
y_train =  macro['S&P 500']



X_train = np.reshape(X_train,(303,1,123))


model = Sequential([layers.Input((1,123)),
                    layers.LSTM(64),
                    layers.Dense(32, activation='relu'),
                    Dropout(0.2),
                    layers.Dense(32, activation='relu'),
                    Dropout(0.2),
                    layers.Dense(1),
                ])

model.compile(optimizer='adam', loss='mse')
model.summary()


model.fit(X_train,y_train)
importances = analyze_feature_importance(model, X_train)
top_10 = np.argsort(importances)[::-1][:10]
top_30 = np.argsort(importances)[::-1][:30]
top_50 = np.argsort(importances)[::-1][:50]

col_10 = macro.columns[top_10]
macro[col_10].to_csv("/content/drive/MyDrive/계특 프로젝트/차원 축소 데이터/macro/LSTM_macro_10.csv")

col_30 = macro.columns[top_30]
macro[col_30].to_csv("/content/drive/MyDrive/계특 프로젝트/차원 축소 데이터/macro/LSTM_macro_30.csv")

col_50 = macro.columns[top_50]
macro[col_50].to_csv("/content/drive/MyDrive/계특 프로젝트/차원 축소 데이터/macro/LSTM_macro_50.csv")
```
